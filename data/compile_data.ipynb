## Insert Holiday (1) or Not Holiday (0)

Washington D.C. public holiday extracted from http://dchr.dc.gov/page/holiday-schedule

```python
dholi = pd.read_csv('raw/holiday_raw.csv')
```

```python
dholi['date'] = pd.to_datetime(dholi['date']).dt.date
```

```python
hd = []

for i in dholi['date']:
    hd.append(str(i))
```

```python
holiday = []
for j in dsb['date']:
    if str(j) in hd:
        holiday.append(1)
    else:
        holiday.append(0)
```

```python
# insert list holiday ke dsb
dsb.insert(1, 'holiday', holiday)
```

## Insert Workingday (1) / Not Workingday (0)

Workingday: Monday to Friday

```python
import datetime
import calendar
workingday = []

for i in dsb['date']:
    print(calendar.day_name[datetime.datetime.strptime(str(i), '%Y-%m-%d').weekday()], datetime.datetime.strptime(str(i), '%Y-%m-%d').weekday())
    if datetime.datetime.strptime(str(i), '%Y-%m-%d').weekday() > 4:
        workingday.append(0)
    else:
        workingday.append(1)
```

```python
dsb.insert(2, 'workingday', workingday)
```

```python
dsb.head()
```

## Insert Season

Season based on meteorological seasons<br>

1: Spring (started at March 1)<br>
2: Summer (started at June 1)<br>
3: Fall (started at September 1)<br>
4: Winter (started at December 1)<br>

source: https://www.ncei.noaa.gov/news/meteorological-versus-astronomical-seasons

```python
spring = datetime.date(2019,3,1)
spring = spring.strftime('%m-%d')

summer = datetime.date(2019,6,1)
summer = summer.strftime('%m-%d')

fall = datetime.date(2019,9,1)
fall = fall.strftime('%m-%d')

winter = datetime.date(2019,12,1)
winter = winter.strftime('%m-%d')

season = []

for i in dsb['date']:
    hari = datetime.datetime.strptime(str(i), '%Y-%m-%d').date()
    hari = hari.strftime('%m-%d')
    if hari >= spring and hari < summer:
        season.append(1)
    elif hari >= summer and hari < fall:
        season.append(2)
    elif hari >= fall and hari < winter:
        season.append(3)
    elif hari < spring or hari >= winter:
        season.append(4)
    
#     print(hari)

len(season)
```

```python
dsb.insert(3, 'season', season)
```

```python
dsb.head()
```

## Insert Weather (temperature, wind speed, humidity)

### Split date to new features (month & year)
