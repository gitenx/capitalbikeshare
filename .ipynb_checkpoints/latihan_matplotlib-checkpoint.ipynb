```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

### bikin plot-nya

```python
plt.plot(x,y,'b')
```

```python
plt.figure(figsize=(15,8)) # lebar, tinggi
plt.plot(x,y,'m--')
plt.plot(y,x,'b')
plt.xlabel('x axis', fontsize=14)
plt.ylabel('y axis', fontsize=14)
plt.title('latihan matplotlib', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.subplot(2,2,1)
plt.plot(y,x, 'b--')
plt.subplot(2,2,2)
plt.plot(x,y, 'r*')
plt.subplot(2,2,3)
plt.plot(y,x, 'g-')
plt.subplot(2,2,4)
plt.plot(x,y, 'k*-')
```

### Matplotlib Object Oriented Method

```python
fig = plt.figure()

axes = fig.add_axes([0, 0, 0.4, 0.4]) # add_axes(left,right,width, height)

axes.plot(y,x, 'y')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('xy')

axes2 = fig.add_axes([0, 0, 0.3, 0.3])

axes2.plot(x,y, 'g')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_axis_off()
```

### subplots

```python
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(10,10))

axs[0][0].plot(y,x,'b--')
axs[1][1].plot(x,y,'r') # axs[row][col]
```

## Yahoo! Historical

```python
from yahoo_historical import Fetcher
import pandas as pd
```

```python
dfmandiri = Fetcher('BMRI.JK', [2008,1,1], [2019,9,10]).getHistorical()
```

```python
dfmandiri.head()
```

```python
dfmandiri.info()
```

```python
dfmandiri.isna().any()
```

```python
plt.plot(dfmandiri['High'], 'g')
```

```python
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(14,11))

axs[0][0].plot(dfmandiri['Open'],'b')
axs[0][0].set_title('BMRI.JK Open')
axs[0][0].set_xlabel('Year')
axs[0][0].set_ylabel('Rupiah')

axs[0][1].plot(dfmandiri['High'],'g')
axs[0][1].set_title('BMRI.JK High')
axs[0][1].set_xlabel('Year')
axs[0][1].set_ylabel('Rupiah')

axs[1][0].plot(dfmandiri['Volume'],'y')
axs[1][0].set_title('BMRI.JK Volume')
axs[1][0].set_xlabel('Year')
axs[1][0].set_ylabel('Rupiah')

axs[1][1].plot(dfmandiri['Adj Close'],'c')
axs[1][1].set_title('BMRI.JK Adj Close')
axs[1][1].set_xlabel('Year')
axs[1][1].set_ylabel('Rupiah')
```
